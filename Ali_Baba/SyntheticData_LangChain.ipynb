```python
from langchain.prompts import FewShotPromptTemplate, PromptTemplate
from langchain.chat_models import ChatOpenAI
from langchain.pydantic_v1 import BaseModel
from langchain_experimental.tabular_synthetic_data.base import SyntheticDataGenerator
from langchain_experimental.tabular_synthetic_data.openai import create_openai_data_generator, OPENAI_TEMPLATE
from langchain_experimental.tabular_synthetic_data.prompts import SYNTHETIC_FEW_SHOT_SUFFIX, SYNTHETIC_FEW_SHOT_PREFIX
```

```python
from pydantic import BaseModel, Field
from typing import Literal

class InsuranceClaimApproval(BaseModel):
    ic_number: int = Field(
        ..., 
        description="Identification number of the claimant; used to verify identity and check historical claim patterns for fraud detection or policy status verification."
    )

    age: int = Field(
        ..., 
        ge=16, le=100, 
        description="Age of the driver; younger or older drivers may have higher risk profiles, influencing claim approval likelihood and coverage eligibility."
    )

    months_as_customer: int = Field(
        ..., 
        ge=0, 
        description="Number of months the individual has been a customer; long-standing customers may be viewed as more trustworthy, potentially influencing approval and coverage favorably."
    )

    vehicle_age_years: int = Field(
        ..., 
        ge=0, le=30, 
        description="Age of the vehicle in years; older vehicles may receive lower payouts due to depreciation, and unusually high values for older cars could raise fraud suspicion."
    )

    vehicle_make: str = Field(
        ..., 
        description="Make of the vehicle; certain models may have higher repair costs or theft risk, affecting claim cost and fraud risk evaluation."
    )

    policy_expired_flag: Literal[0, 1] = Field(
        ..., 
        description="1 if the policy has expired; 0 otherwise. An expired policy typically leads to automatic claim denial unless special clauses apply."
    )

    deductible_amount: float = Field(
        ..., 
        ge=0, 
        description="Deductible amount in the policy; affects how much the insurer pays out. Very low deductibles with high repair claims may raise fraud flags."
    )

    repair_amount: float = Field(
        ..., 
        ge=0, 
        description="Estimated repair cost; high values relative to market value or damage severity may suggest fraudulent exaggeration or impact approval if excessive."
    )

    market_value: float = Field(
        ..., 
        ge=0, 
        description="Market value of the vehicle; helps cap the maximum payout. Claims that exceed this may be denied or flagged for investigation."
    )

    damage_severity_score: float = Field(
        ..., 
        ge=0, le=1.0, 
        description="Normalized severity score of damage from 0 to 1; high severity generally leads to higher payouts, but inconsistencies with repair cost or accident reports could indicate fraud."
    )

    at_fault_flag: Literal[0, 1] = Field(
        ..., 
        description="1 if the insured is at fault; 0 otherwise. Being at fault may reduce or deny claim approval, depending on the policy coverage."
    )

    time_to_report_days: int = Field(
        ..., 
        ge=0, 
        description="Days taken to report the claim; long delays may reduce credibility or indicate fraud. Prompt reports are generally more trustworthy."
    )

    claim_reported_to_police_flag: Literal[0, 1] = Field(
        ..., 
        description="1 if the claim was reported to police; 0 otherwise. Police involvement typically improves claim credibility, especially for serious incidents or third-party involvement."
    )

    license_type_missing_flag: Literal[0, 1] = Field(
        ..., 
        description="1 if license type information is missing; 0 otherwise. Missing license data may delay or deny claims and is a potential fraud signal."
    )

    num_third_parties: int = Field(
        ..., 
        ge=0, 
        description="Number of third parties involved in the claim; higher counts increase complexity and potential cost, and inconsistencies may indicate fraud."
    )

    num_witnesses: int = Field(
        ..., 
        ge=0, 
        description="Number of witnesses for the incident; more witnesses typically strengthen the claim's credibility and support approval."
    )

    approval_flag: Literal[0, 1] = Field(
        ..., 
        description="Target variable: 1 if claim is approved; 0 if declined. This is the predicted or labeled outcome for machine learning classification models."
    )
```

```python
# Helper function to format each example into a string
def format_example(example_dict):
    return ", ".join([f"{key.replace('_', ' ').title()}: {value}" for key, value in example_dict.items()])

# Convert raw examples to FewShotPromptTemplate format
examples = [{"example": format_example(example)} for example in raw_claimapproval_examples]

# Set up the prompt template
OPENAI_TEMPLATE = PromptTemplate(input_variables=["example"], template="{example}")
```

```python
SYNTHETIC_FEW_SHOT_PREFIX = "Here are some examples of insurance claims data in Malaysia:\n"
SYNTHETIC_FEW_SHOT_SUFFIX = "\nBased on the above examples, generate a new insurance claim with the following details:\nSubject: {subject}\nExtra Information: {extra}"

prompt_template = FewShotPromptTemplate(
    prefix=SYNTHETIC_FEW_SHOT_PREFIX,
    examples=examples,
    suffix=SYNTHETIC_FEW_SHOT_SUFFIX,
    input_variables=["subject", "extra"],
    example_prompt=OPENAI_TEMPLATE,
    example_separator="\n\n"
)
```

```python
# Create the synthetic data generator
synthetic_data_generator = create_openai_data_generator(
    output_schema=InsuranceClaimApproval,
    llm=ChatOpenAI(temperature=1),
    prompt=prompt_template,
)
```

```python
# Generate synthetic insurance claims
synthetic_results = synthetic_data_generator.generate(
    subject="insurance_claim_approval",
    extra="Ensure the IC number follows the Malaysian NRIC format and use uncommon Malaysian names.",
    runs=50,
)
```
