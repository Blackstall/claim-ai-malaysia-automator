```python
import fitz  # PyMuPDF
doc = fitz.open("insurance_policy_company_A.docx")
full_html = ""
for page in doc:
    # get HTML-formatted content to preserve structure
    full_html += page.get_text("html")
```

```python
from markdownify import markdownify as md

markdown_content = md(full_html)
```

```python
import re

def normalize_markdown(text: str) -> str:
    import re

    # Convert literal '\\n' to real newlines, just in case
    text = text.replace('\\n', '\n')

    # Split the text into paragraph blocks (separated by 2+ newlines)
    blocks = re.split(r'\n{2,}', text)

    normalized_blocks = []
    for block in blocks:
        lines = block.splitlines()
        # If it's a list, heading, or table block — keep as is
        if all(re.match(r'^(\s*[-*•]|\d+\.|\|)', line) or line.strip() == '' for line in lines):
            normalized_blocks.append('\n'.join(lines))
        else:
            # Rejoin broken lines inside normal text blocks
            joined = ' '.join(line.strip() for line in lines)
            normalized_blocks.append(joined)

    # Join blocks with 2 newlines again
    return '\n\n'.join(normalized_blocks)


# Assuming markdown_content is your decoded markdown string:
markdown_clean = normalize_markdown(markdown_content)
markdown_clean
```

```python
# Step 3: Split markdown into chunks
def split_into_chunks(text, max_tokens=300):
    import textwrap
    return textwrap.wrap(text, width=max_tokens)

chunks = split_into_chunks(markdown_clean)
```

```python
from openai import OpenAI
import os
from tqdm import tqdm

# Setup DashScope OpenAI-compatible client
client = OpenAI(
    api_key="sk-91d59f4cf5ff44cf9f8280a91dacbb1c",
    base_url="https://dashscope-intl.aliyuncs.com/compatible-mode/v1"
)

# Function to split a list into batches of max size `n`
def batch_list(lst, n=10):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]

# Process in batches of max 10
all_embeddings = []
for chunk_batch in tqdm(batch_list(chunks, 10), desc="Embedding batches"):
    response = client.embeddings.create(
        model="text-embedding-v3",
        input=chunk_batch,
        encoding_format="float"  # Remove `dimensions=1024` – not supported by DashScope
    )
    all_embeddings.extend([e.embedding for e in response.data])

# Now you have:
# - `chunks`: your input texts
# - `all_embeddings`: matching list of embeddings
```
